```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\kunal\OneDrive\Desktop\practice\Projects\Suicide rates\suicide_rates_1990-2022.csv')
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=1, thresh=int(0.3 * len(df)))
```

```python
df['GDP'] = df['GDP'].fillna(df['GDP'].mean())
df['GDPPerCapita'] = df['GDPPerCapita'].fillna(df['GDPPerCapita'].mean())
```

```python
df["CauseSpecificDeathPercentage"] = df["CauseSpecificDeathPercentage"].fillna(df["CauseSpecificDeathPercentage"].median())
df["DeathRatePer100K"] = df["DeathRatePer100K"].fillna(df["DeathRatePer100K"].mean())
```

```python
df["Population"] = df["Population"].interpolate(method="linear")
```

```python
df['Generation'] = df['Generation'].fillna(df['Generation'].mode())
df["SuicideCount"] = df["SuicideCount"].fillna(0)
df["GrossNationalIncome"] = df["GrossNationalIncome"].fillna(method='ffill')
df["GNIPerCapita"] = df["GNIPerCapita"].fillna(method='bfill')

```

```python
df["EmploymentPopulationRatio"] = df["EmploymentPopulationRatio"].fillna(method="bfill")
df = df.drop(columns=["InflationRate"])
```
